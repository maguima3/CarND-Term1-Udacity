```python
import csv
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline


data_path = '../../../../CarND-P3-data/'
```

```python
lines = []
with open(data_path + 'driving_log.csv', 'r') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)
        for line in reader:
                lines.append(line)
print(lines[7])
```

```python
angles = []
for line in lines:
    angles.append(float(line[3]))

plt.plot(range(len(angles)), angles)
plt.show()

positive = len(sum(np.where(np.asarray(angles) > 0)))
print('pos', positive)

negative = len(sum(np.where(np.asarray(angles) < 0)))
print('neg', negative)

total = len(angles)
print('total', total)

print('pos%%', positive/total)
print('neg%%', negative/total)

```

```python
from random import randint, uniform
import math 



def random_flip(image, angle):
        if (randint(0,1)):
                return np.fliplr(image), -angle
        else:
                return image, angle   

def random_h_translation(image, angle):
    h_translation_range = 25
    translation = uniform(-h_translation_range, h_translation_range)
    print('h_translation', translation)
    transformation = np.float32([[1,0,translation],[0,1,0]])

    rows, cols = image.shape[0], image.shape[1]
    image_translated = cv2.warpAffine(image, transformation, (cols,rows))
    angle_corrected = angle + (translation/h_translation_range) * 0.1

    return image_translated, angle_corrected, translation


def random_v_translation(image, angle):
        
    v_translation_range = 10
    translation = uniform(-v_translation_range, v_translation_range)
    print('v_translation', translation)

    transformation = np.float32([[1,0,0],[0,1,translation]])
    rows, cols = image.shape[0], image.shape[1]
    image_translated = cv2.warpAffine(image, transformation, (cols, rows))
    angle_corrected = angle + (translation/v_translation_range) * 0.05
    return image_translated, angle_corrected, translation


def preprocess_input(image):
    processed = image[60:140, :]
    processed = cv2.resize(processed, (200, 66))
    processed = cv2.cvtColor(processed, cv2.COLOR_RGB2YUV)
    return processed


def crop(image, h_translation, v_translation):
    shape = image.shape
    rows, cols = image.shape[0], image.shape[1]

    if h_translation >= 0:
    	h_init = int(math.ceil(h_translation))
    	h_end = cols
    else:
    	h_init = 0
    	h_end = cols + int(math.floor(h_translation))

    if v_translation >= 0:
    	v_init = int(math.ceil(v_translation))
    	v_end = rows
    else:
    	v_init = 0
    	v_end = rows + int(math.floor(v_translation))

    image = image[v_init:v_end, h_init:h_end]
    image = cv2.resize(image, (200,66), interpolation = cv2.INTER_LINEAR)
    return image
```

```python
print('Original')
line = lines[2]        
filename = line[0].split('/')[-1]
current_path = data_path + 'IMG/' + filename
img_test = cv2.imread(current_path)
img_test = img_test[60:140, :]
img_test = cv2.resize(img_test, (200, 66), interpolation = cv2.INTER_AREA)
ang_test =  float(line[3])
plt.imshow(img_test)
plt.title(ang_test)
plt.axis('off')
plt.show()


print('Translated image')
(img1, ang1, h_translation) = random_h_translation(img_test, ang_test)
(img2, ang2, v_translation) = random_v_translation(img1, ang1)
img3 = crop(img2, h_translation, v_translation)

plt.imshow(img3)
plt.title(ang2)
plt.axis('off')
plt.show()

print('Flipped image')
img4, ang4 = random_flip(img_test, ang_test)
plt.imshow(img4)
plt.title(ang4)
plt.axis('off')
plt.show()
```
